```python
# BAR GRAPH
import matplotlib.pyplot as plt 

students = ['Mobin','Faisal','chandan', 'ethan', 'sahil']
marks = [80, 85, 75, 85, 90]

plt.bar(students,marks)

# label
plt.xlabel("STUDENTS")
plt.ylabel("MARKS")
plt.title("MARKS OF THE STUDENTS")

plt.show()
```

```python
import matplotlib.pyplot as plt

states = ['Goa','kerala','delhi','punjab']
traffics = [40,50,70,28]

plt.bar(states,traffics)

plt.xlabel('STATES')
plt.ylabel('TRAFFIC')
plt.title('WHICH STATES HAVING MORE TRAFFIC')

plt.show()
```
